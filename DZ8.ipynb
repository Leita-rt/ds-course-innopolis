```python
import numpy as np
import pandas as pda
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
array = np.genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                      dtype=[('sepal_length','f8'),('sepal_width','f8'),('petal_length','f8'),('petal_width','f8'),('class','U15')],
                      delimiter=',')
```

```python
array = pd.DataFrame(array)
```

```python
array.describe().round(2)
```

```python
def plot_ax(ax, column, title):
    ax.hist(column)
    ax.set_title(title)
```

```python
def plot(array):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

    plot_ax(ax1, array.sepal_length, 'Гистограмма sepal_length')

    plot_ax(ax2, array.sepal_width, 'Гистограмма sepal_width')

    plot_ax(ax3, array.petal_length, 'Гистограмма petal_length')

    plot_ax(ax4, array.petal_width, 'Гистограмма petal_width')
```

```python
plot(array)
```

## Вычисление статистических величин для класса Iris-setosa, а также построение гистрограмм для этого класса:

```python
class1 = array.groupby('class').get_group('Iris-setosa')
print('Статистические величины для класса Iris-setosa:\n\n', class1.describe().round(2))
```

```python
plot(class1)
```

## Вычисление статистических величин для класса Iris-versicolor, а также построение гистрограмм для этого класса:

```python
plot(class2)
```
